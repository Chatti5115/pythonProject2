```
import requests
```

```
# key = '8834005068'
# type = 'json'
# iYear = '2020' 
# StartUnitCode = '101'
# EndUnitCode = '105'

URL = 'http://data.ex.co.kr/openapi/specialAnal/trafficFlowByTime?key=8834005068&type=json&iStdYear=2018'
# url = URL + key + '&type=json'+'&iYear='+iYear
```

```
response = requests.get(URL)
```

```
response
```

```
json = response.json()
```

```
json
```

```
cars = json['trafficFlowByTimeLists']
```

```
cars
```

```
records = []
for car in cars:
    dic = {}
    dic['date'] = car['stdYear']
    dic['time'] = car['stdHour']
    dic['type'] = car['sphlDfttNm']
    records.append(dic)
records
```
